```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
from datetime import datetime
```

```python
pd.set_option('display.max_rows',30)
df = pd.read_csv('glassdoor.csv', parse_dates=['Founded'])
df
```

```python
#how many rows are in the orginal df
len(df)
```

```python
#let's look at the unique columns 
df.columns
```

```python
#let's drop the 1st unnamed column
df = df.drop('Unnamed: 0', axis=1)
df.head()
```

```python
df.info() #from this info we can see that there aren't really null values 
          #but there are missing values for sure 
```

```python
df.describe() #here we can see that there are ratings of -1
              #which really shouldn't be the case!
              #we have to fix this 
```

```python
#create a new column that is true if the job pays hourly 
df['hourly'] = df['Salary Estimate'].apply(lambda x: 1 if 'per hour' in x.lower() else 0)

#df[df['hourly'] == 1]
```

```python
#create a column that is true if the sal is employer provided
df['Employer Provided Sal'] = df['Salary Estimate'].apply(lambda x: 1 if 'employer provided' in x.lower() else 0)

#df[df['Employer Provided Sal'] == 1]
```

```python
#remove salary estimates with missing values 
#that is the value is '-1'

df = (df[df['Salary Estimate'] != '-1'])
len(df) #rows have reduced in number to 742 from 956
```

```python
#rep the sal est as ints with min and max vals 
df['Salary Estimate'].unique()
```

```python
df['Salary Estimate'] = df['Salary Estimate'].str.split('(').str[0]
sal = df.copy()
sal
```

```python
sal['Salary Estimate'].unique()
```

```python
#let's remove the phrase 'Per Hour'
sal['Salary Estimate'] = sal['Salary Estimate'].str.replace('Per Hour','').str.replace('Employer Provided Salary:','')
sal['Salary Estimate'].unique()
```

```python
#let's remove whitespace on either side
sal['Salary Estimate'] = sal['Salary Estimate'].str.strip()
sal['Salary Estimate'].unique() #all whitespace stripped
```

```python
#finally let's have the est be a range of ints
sal['Salary Estimate'] = sal['Salary Estimate'].str.replace('$','').str.replace('K','')
sal['Salary Estimate'].unique() 
```

```python
sal.head()
```

```python
#create two columns, one that will have the min and the other the max, salary
sal['min_sal'] = sal['Salary Estimate'].str.split('-').str[0].astype(np.int64)
sal.head()
```

```python
sal['max_sal'] = sal['Salary Estimate'].str.split('-').str[1].astype(np.int64)
sal.head()
```

```python
sal.info() #let's see if the new columns we made have data type of int64
```

```python
#let's create a column with the avg sal 
#gotten using the min and max sal

sal['avg_sal'] = (sal['min_sal'] + sal['max_sal'])/2
sal['avg_sal'].unique()
```

```python
sal[sal['Rating']>=0] #if there is a rating, the rating comes last after the company name
                      #we need to remove the rating and only have the company name
```

```python
#return the company name only, so do away with the rating 
def name(rating, cname):
    if rating<0:
        return cname
    
    else:
        cname = cname.split()
        cname = cname[:-1]
        cname = ' '.join(cname)
        return cname 

name(3,'The Hanover Insurance Group\r\n3.7')
```

```python
#NB: ONLY RUN THIS CELL ONCE 
#let's apply the function to our sal df
#sal['Company Name'].unique()

sal['Company Name'] = np.vectorize(name)(sal['Rating'], sal['Company Name'])
sal['Company Name'].unique()
```

```python
#let's create another column that contains the state where the job/company is 
sal['State'] = sal['Location'].str.split(',').str[-1]
sal[['Location', 'State']]
```
